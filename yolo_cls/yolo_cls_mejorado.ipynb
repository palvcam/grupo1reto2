# Evaluación

```python
TEST_DIR = DATASET_PATH / "test"

MODEL_PATH = PATH / "../yolo_cls/runs/classify/train/weights/best.pt"
model = YOLO(MODEL_PATH)
```

```python
results = []
for cls_dir in TEST_DIR.iterdir():
    if cls_dir.is_dir():
        print(f"Procesando carpeta: {cls_dir.name}")
        results.extend(model(cls_dir))
```

```python
# Predicciones
for r in results:
    probs = r.probs.data.cpu().numpy()
    names_dict = r.names

    pred_class_id = np.argmax(probs)
    pred_class_name = names_dict[pred_class_id]
    confidence = probs[pred_class_id]

    print(f"{r.path} → {pred_class_name} ({confidence:.4f})")
```

```python
correct = 0
total = 0
for r in results:
    probs = r.probs.data.cpu().numpy()
    pred_class = r.names[np.argmax(probs)]
    true_class = Path(r.path).parent.name

    if pred_class == true_class:
        correct += 1
    total += 1

accuracy = correct / total if total > 0 else 0.0
print("\n--- RESULTADOS FINALES ---")
print(f"Total imágenes test: {total}")
print(f"Correctas: {correct}")
print(f"Accuracy test: {accuracy:.4f}")
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np

# Crear listas para almacenar resultados
y_true = []
y_pred = []

for r in results:
    probs = r.probs.data.cpu().numpy()
    pred_class = r.names[np.argmax(probs)]
    true_class = Path(r.path).parent.name

    y_true.append(true_class)
    y_pred.append(pred_class)

# Guardar resultados en CSV
df_results = pd.DataFrame({
    "image_path": [r.path for r in results],
    "true_class": y_true,
    "pred_class": y_pred,
})
df_results.to_csv("classification_results.csv", index=False)

# Métricas
print("=== Classification Report ===")
print(classification_report(y_true, y_pred, digits=4))

# Matriz de confusión
cm = confusion_matrix(y_true, y_pred, labels=list(set(y_true)))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=set(y_true), yticklabels=set(y_true))
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

# Prueba de diferentes hiperparámetros

```python
import matplotlib.pyplot as plt
import itertools
```

```python
PROJECT_DIR = Path(".").resolve()
```

```python
# Definimos hiperparámetros a probar
# Ejemplo: batch size, learning rate, epochs
batch_sizes = [8, 16]
learning_rates = [0.001, 0.005]
epochs = [10, 20]  # para prueba rápida, puedes aumentar para entrenamientos serios
```

```python
# Creamos combinaciones
param_combinations = list(itertools.product(batch_sizes, learning_rates, epochs))
```

```python
# CSV donde guardaremos resultados
RESULTS_CSV = PROJECT_DIR / "yolo_hyperparam_results.csv"
results = []
```

```python
# Bucle de entrenamiento
for batch_size, lr, ep in param_combinations:
    print(f"\n=== Entrenando con batch_size={batch_size}, lr={lr}, epochs={ep} ===")

    model = YOLO("yolo11s-cls.pt")  # cargamos modelo base
    # Entrenamos
    model.train(
        data=DATASET_PATH,
        epochs=ep,
        batch=batch_size,
        lr0=lr,
        verbose=False
    )

    # Evaluamos
    metrics = model.val()  # devuelve ClassifyMetrics

    # Extraemos métricas de clasificación
    top1 = metrics.top1
    top5 = metrics.top5

    # Guardamos resultados
    results.append({
        'batch_size': batch_size,
        'lr': lr,
        'epochs': ep,
        'top1': top1,
        'top5': top5
    })

    print(f"Resultados: Top-1={top1:.3f}, Top-5={top5:.3f}")
```

```python
# Guardamos CSV
df = pd.DataFrame(results)
df.to_csv(RESULTS_CSV, index=False)
print(f"\nCSV guardado en {RESULTS_CSV}")
```

```python
from ultralytics import YOLO
from pathlib import Path
import pandas as pd

# Carpeta con imágenes de validación
TEST_DIR = Path("test")

# Cargar el modelo final
model = YOLO('./runs/classify/train10/weights/best.pt')   # tu modelo entrenado

# Lista para guardar resultados
predictions = []

# Predecir cada imagen en la carpeta
for img_path in TEST_DIR.glob("*.*"):  # toma todos los archivos
    results = model.predict(img_path)

    # Top-1
    top1_class = results.labels[0]   # clase predicha
    top1_prob = results.probs[0]     # probabilidad de la clase predicha

    # Guardar
    predictions.append({
        "image": img_path.name,
        "top1_class": top1_class,
        "top1_prob": top1_prob
    })

    # Mostrar la imagen con la predicción
    results.show()

# Guardar todas las predicciones en un CSV
df_pred = pd.DataFrame(predictions)
df_pred.to_csv("predicciones_test.csv", index=False)
print("Predicciones guardadas en predicciones_test.csv")
```
