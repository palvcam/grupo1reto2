```python
import cv2
import shutil
from pathlib import Path
import numpy as np

# --- RUTAS ---
BASE_DIR = Path("data_imagenes")
RAW_DIR = BASE_DIR / "kos01"

DEFECT_DIR = BASE_DIR / "defect"
OK_DIR = BASE_DIR / "ok"

DEFECT_DIR.mkdir(exist_ok=True)
OK_DIR.mkdir(exist_ok=True)

# --- PARÁMETROS DE DETECCIÓN ---
WHITE_THRESHOLD = 10      # umbral para considerar un píxel blanco
MIN_WHITE_PIXELS = 50     # mínimo de píxeles blancos para decir "defecto"

# --- FUNCIÓN PARA DETECTAR DEFECTO ---
def has_defect(label_path):
    label = cv2.imread(str(label_path), cv2.IMREAD_GRAYSCALE)

    if label is None:
        raise ValueError(f"No se pudo leer {label_path}")

    white_pixels = np.sum(label > WHITE_THRESHOLD)
    return white_pixels > MIN_WHITE_PIXELS


# --- PROCESO PRINCIPAL ---
for img_path in RAW_DIR.glob("Part*.jpg"):

    label_path = img_path.with_name(img_path.stem + "_label.bmp")

    if not label_path.exists():
        print(f"Falta label para {img_path.name}")
        continue

    if has_defect(label_path):
        target_dir = DEFECT_DIR
        status = "DEFECT"
    else:
        target_dir = OK_DIR
        status = "OK"

    # Copiar imagen y su pareja
    shutil.copy(img_path, target_dir / img_path.name)
    shutil.copy(label_path, target_dir / label_path.name)

    print(f"{status}: {img_path.name}")
```

```python
from pathlib import Path
import cv2
import numpy as np
import shutil

# --- RUTAS ---
BASE_DIR = Path("data_imagenes")

OUTPUT_DIR = BASE_DIR / "clasificado"
DEFECT_DIR = OUTPUT_DIR / "defecto"
OK_DIR = OUTPUT_DIR / "no_defecto"

DEFECT_DIR.mkdir(parents=True, exist_ok=True)
OK_DIR.mkdir(parents=True, exist_ok=True)

# --- PARÁMETROS ---
WHITE_THRESHOLD = 10
MIN_WHITE_PIXELS = 50

# --- FUNCIÓN DE DETECCIÓN ---
def has_defect(label_path):
    label = cv2.imread(str(label_path), cv2.IMREAD_GRAYSCALE)

    if label is None:
        raise ValueError(f"No se pudo leer {label_path}")

    white_pixels = np.sum(label > WHITE_THRESHOLD)
    return white_pixels > MIN_WHITE_PIXELS


# --- PROCESO PRINCIPAL ---
for i in range(1, 51):
    raw_dir = BASE_DIR / f"kos{i:02d}"

    if not raw_dir.exists():
        print(f"Carpeta no encontrada: {raw_dir}")
        continue

    kos_name = raw_dir.name  # kos01, kos02, ...

    print(f"\nProcesando {kos_name}...")

    for img_path in raw_dir.glob("Part*.jpg"):

        label_path = img_path.with_name(img_path.stem + "_label.bmp")

        if not label_path.exists():
            print(f"Falta label para {img_path.name}")
            continue

        if has_defect(label_path):
            target_dir = DEFECT_DIR
            status = "DEFECTO"
        else:
            target_dir = OK_DIR
            status = "NO_DEFECTO"

        # nombre único: kosXX_PartY.jpg
        new_name = f"{kos_name}_{img_path.name}"
        shutil.copy(img_path, target_dir / new_name)

        print(f"{status}: {new_name}")
```

# YOLO CLS

```python
import random
import shutil

# --- RUTAS ---
BASE_DIR = Path("data_imagenes/clasificado")
OUTPUT_DIR = Path("dataset_cls")

CLASSES = ["defecto", "no_defecto"]

SPLITS = {
    "train": 0.7,
    "val": 0.15,
    "test": 0.15
}

random.seed(42)  # reproducible

# --- CREAR ESTRUCTURA ---
for split in SPLITS:
    for cls in CLASSES:
        (OUTPUT_DIR / split / cls).mkdir(parents=True, exist_ok=True)

# --- PROCESO ---
for cls in CLASSES:
    images = list((BASE_DIR / cls).glob("*.jpg"))
    random.shuffle(images)

    n_total = len(images)
    n_train = int(SPLITS["train"] * n_total)
    n_val = int(SPLITS["val"] * n_total)

    splits = {
        "train": images[:n_train],
        "val": images[n_train:n_train + n_val],
        "test": images[n_train + n_val:]
    }

    for split, imgs in splits.items():
        for img in imgs:
            shutil.copy(
                img,
                OUTPUT_DIR / split / cls / img.name
            )

    print(f"{cls}: {n_total} imágenes → "
          f"train={len(splits['train'])}, "
          f"val={len(splits['val'])}, "
          f"test={len(splits['test'])}")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt')  # load a pretrained model (recommended for training)

model.train(data=PATH / "dataset_cls",
            epochs=10, imgsz=64)
```

```python
# Modelo entrenado
MODEL_PATH = PATH / "runs" / "classify" / "train" / "weights" / "best.pt"

# Dataset de test
TEST_DIR = PATH / "dataset_cls" / "test"

# =========================================================
# CARGA DEL MODELO
# =========================================================

model = YOLO(MODEL_PATH)

# =========================================================
# INFERENCIA SOBRE TODO EL TEST (recorriendo clases)
# =========================================================

results = []

for cls_dir in TEST_DIR.iterdir():
    if cls_dir.is_dir():
        print(f"Procesando carpeta: {cls_dir.name}")
        results.extend(model(cls_dir))

# =========================================================
# MOSTRAR PREDICCIONES IMAGEN A IMAGEN
# =========================================================

print("\n--- PREDICCIONES ---\n")

for r in results:
    probs = r.probs.data.cpu().numpy()
    names_dict = r.names

    pred_class_id = np.argmax(probs)
    pred_class_name = names_dict[pred_class_id]
    confidence = probs[pred_class_id]

    print(f"{r.path} → {pred_class_name} ({confidence:.4f})")

# =========================================================
# EVALUACIÓN: ACCURACY EN TEST
# =========================================================

correct = 0
total = 0

for r in results:
    probs = r.probs.data.cpu().numpy()
    pred_class = r.names[np.argmax(probs)]
    true_class = Path(r.path).parent.name  # carpeta = etiqueta real

    if pred_class == true_class:
        correct += 1
    total += 1

accuracy = correct / total if total > 0 else 0.0

print("\n--- RESULTADOS FINALES ---")
print(f"Total imágenes test: {total}")
print(f"Correctas: {correct}")
print(f"Accuracy test: {accuracy:.4f}")
```
