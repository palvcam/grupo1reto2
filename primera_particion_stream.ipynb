```python
from pathlib import Path
import shutil
import random

# =========================
# CONFIG
# =========================
SRC_DIR = Path("Imagenes_defectos")
STREAM_DIR = Path("Ingesta") / "stream_pool"

IMAGES_PER_KOS = 2      # cuántas imágenes mover por kosXX
RANDOM_SEED = 42

random.seed(RANDOM_SEED)

# =========================
# PREPARAR DESTINO
# =========================
STREAM_DIR.mkdir(parents=True, exist_ok=True)

# =========================
# PROCESO
# =========================
for kos_dir in sorted(SRC_DIR.glob("kos*")):
    if not kos_dir.is_dir():
        continue

    images = sorted(kos_dir.glob("Part*.jpg"))
    if not images:
        continue

    n_select = min(IMAGES_PER_KOS, len(images))
    selected = random.sample(images, n_select)

    print(f"{kos_dir.name}: moviendo {n_select}/{len(images)} imágenes al stream")

    # crear subcarpeta stream
    dst_kos = STREAM_DIR / kos_dir.name
    dst_kos.mkdir(parents=True, exist_ok=True)

    for img_path in selected:
        # mover imagen
        dst_img = dst_kos / img_path.name
        shutil.move(img_path, dst_img)

        # mover label si existe
        label_path = img_path.with_name(f"{img_path.stem}_label.bmp")
        if label_path.exists():
            shutil.move(label_path, dst_kos / label_path.name)

print("\n✔ Imágenes de stream separadas correctamente.")
print("✔ Imagenes_defectos ahora solo contiene datos de entrenamiento.")
print("✔ Ingesta/stream_pool contiene datos nunca vistos.")
```
